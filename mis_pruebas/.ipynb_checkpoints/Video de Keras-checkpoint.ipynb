## Activation Functions

In an artificial neural network, the activation function of a neuron defines the output of that neuron fiven a set of inputs.
- Biologically inspired by activity in our brains, where different neurons fire, or are activated by different stimuli

```python
from keras.models import Sequential
from keras.layers import Dense, Activation

model = Sequential([
    Dense(5, input_shape=(3,), activation='relu')
])
```

```python
# Another way of doing the same:

model = Sequential()
model.add(Dense(5, input_shape=(3,)))
model.add(Activation('relu')) # Activation layer
```

## Training

Training a model is solving an optimization problem. What we are trying to optimize is the weights
- Optimizing weights (depending of the optimation algorithm that we are using (optimizer).
    * with Stochastic Gradient Descent (SGD)
- Objective: Minimize the loss function (bring that function as close to 0 as possible).
  This loss function is the 1/2 * (y - y_hat) ^ 2

## Learning

"Note: the gradient is just the derivative of a function with several arguments"

For the model to learn just means to calculate the weights.
At first, we have arbitrary weights. When the first input training passes, we calculate the error of that prediction by looking at what the model predicted for that input vs what the input's true label is. 
Then we calculate the gradient of the loss function within each of the weights that have been set. 

Once we have the value of the gradient, this value will then be multiplied by something called "a learning rate".
This rate is usually between 0.01 and 0.0001.
Then we add this value to the weight we were calculating.

adjustments = d(loss) / d(weight_i) * learning_rate

```python
import keras
from keras import backend as K
from keras.models import Sequential
from keras.layers import Activation
from keras.layers.core import Dense
from keras.optimizers import Adam
from keras.metrics import categorical_crossentropy
```

```python
# In order to train a model, the first thing we have to do is to build a model.

model = Sequential([
    Dense(16, input_shape=(3,), activation='relu'),
    Dense(32, activation='relu'),
    Dense(2, activation='softmax')
])
# input layer with 1 neuron, 2 hidden layers with 16 and 32 neurons, and 1 output layer with 2 neurons
```

```python
# Once we have a model, the next thing we have to do before we can train it is compile our model

model.compile(Adam(lr=.0001), loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Here Adam is the optimizer, lr is the learning rate, loss is the loss function we are using.
# The metrics is what is printing out when we're training our model we care about seeing.
```
